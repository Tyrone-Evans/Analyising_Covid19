```python
#Outliers
#Detecting Outliers using IQR (interquartile Rannge)
#create sample Dataframe
data = {'Score': [85,90,78,92,88,76,95,89,1000]}
df = pd.DataFrame(data)
Q1 = df['Score'].quantile(0.25)
Q3 = df['Score'].quantile(0.75)

#create a sample
#1 calculate Q1

#Calculate IQR
IQR = Q3 - Q1


#Detect outliers
outliers = df[(df['Score'] < Q1 - 1.5 * IQR) | (df['Score'] > Q3 + 1.5 * IQR)]
print(outliers)
```

```python
import pandas as pd

#load csv
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
df = pd.read_csv(url)

#clean data
df = df.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], var_name='Date', value_name='Confirmed')
df['Date'] = pd.to_datetime(df['Date'])
```

```python
#grouping data
df_grouped = df.groupby(['Country/Region', 'Date']).sum().reset_index()

```
